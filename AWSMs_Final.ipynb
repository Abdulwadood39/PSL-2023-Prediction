```python
from math import ceil

DataFrames = {}
probabilities = {}
for i in df.Year.unique():
    DataFrames[i] = df[df['Year'] == i]
    probabilities[i] = {}
    for j in teams:
        probabilities[i][j] = {} 
        Total_Matches = 0
        for k in teams:
            if j == k:
                continue
            try:
                wins = ceil(DataFrames[i].loc[DataFrames[i].Teams == j][k].values[0]/2)
                losses = ceil(DataFrames[i].loc[DataFrames[i].Teams == k][j].values[0]/2)
                Total_Matches += wins
                Total_Matches += losses
                probabilities[i][j][k] = wins/Total_Matches
            except:
                continue
            
Years = df.Year
X = df.Teams
Y = df.Teams.unique()
Probs = df

for i in range(len(Years)):
    for j in range(len(Y)):
        if X[i] != Y[j]:
            try:
                Probs.loc[i,Y[j]] = float(probabilities[Years[i]][X[i]][Y[j]])
            except:
                Probs.loc[i,Y[j]] = 0
        else:
            Probs.loc[i,Y[j]] = 0
Probs      
```

```python
LQ = []
PZ = []
IU = []
KK = []
MS = []
QG = []

arr = [LQ, PZ, IU, KK, MS, QG]
teams = ['Lahore Qalandars', 
         'Peshawar Zalmi', 
         'Islamabad United', 
         'Karachi Kings', 
         'Multan Sultans', 
         'Quetta Gladiators']

for index, i in enumerate(arr):
    arr[index].extend(Probs[Probs['Year'] == 2016][teams[index]].values*0.105)
    arr[index].extend(Probs[Probs['Year'] == 2017][teams[index]].values*0.115)
    arr[index].extend(Probs[Probs['Year'] == 2018][teams[index]].values*0.125)
    arr[index].extend(Probs[Probs['Year'] == 2019][teams[index]].values*0.14)
    arr[index].extend(Probs[Probs['Year'] == 2020][teams[index]].values*0.15)
    arr[index].extend(Probs[Probs['Year'] == 2021][teams[index]].values*0.175)
    arr[index].extend(Probs[Probs['Year'] == 2022][teams[index]].values*0.175)  
    
    Probs[teams[index]] = arr[index]
```

```python
Agg_Probs = Probs.groupby('Teams').sum().drop('Year', axis = 1)
Agg_Probs
```

## Prediction Function


```python

Home_Win = {
    'Islamabad United':  0.5769230769230769,
    'Karachi Kings' :  0.6086956521739131,
    'Peshawar Zalmi' :  0.6538461538461539,
    'Quetta Gladiators' :  0.56,
    'Lahore Qalandars' :  0.5263157894736842,
    'Multan Sultans' :  0.35714285714285715
}

def Predict(team1,team2,venue):
    x = Agg_Probs.loc[team1][team2]
    y = Agg_Probs.loc[team2][team1]
    if venue in Home_Win.keys():
        if venue in team2 :
            y *= Home_Win[venue]
        else:
            x *= Home_Win[venue]
        
    if x > y:
        return team1
    else:
        return team2

    
```

```python
df = pd.read_csv("./ranking-stage-submission.csv")#.drop('Unnamed: 0', axis = 1)
# replace Islamabad Unites with Islamabad United
df['Team_1'] = df['Team_1'].replace('Islamabad Unites', 'Islamabad United')
df['Team_2'] = df['Team_2'].replace('Islamabad Unites', 'Islamabad United')
# replace Rawalpindi with Islamabad as its Islamabad's Home Ground
df['Venue'] = df['Venue'].replace('Rawalpindi', 'Islamabad')

df
```

## Results

```python
Answers = []
for i in range(len(df)):
    Answers.append(Predict(str(df.iloc[i]['Team_1']),str(df.iloc[i]['Team_2']),str(df.iloc[i]['Venue'])))

df['prediction'] = Answers
df
```
